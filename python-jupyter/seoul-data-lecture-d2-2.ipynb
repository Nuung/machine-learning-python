```python

# ex 4
# columnwise operation
n = 10
p = 5

np.random.seed(1)
x = np.random.normal(size = (n,p))
betavec = np.random.uniform(size = (p,1))
y = x@betavec
print("y >>", y)


z = np.zeros(n)
for i in range(p):
    z += x[:,i]*betavec[i] 
z = z.reshape(-1,1) # -1은 여기에 맞추겠다, 열의 개수를 1로 맞추겠다
print("z >>", z)

# numpy는 index 로 조질 순 없지만, concat 가능
np.concatenate((y,z), axis = 1)
```

```python

# ex 5 - 1
# 셈플 데이터로 연산하기
import pandas as pd
data = pd.read_csv('./datas/sample.csv', encoding='cp949')
n = data.shape

data
```

```python

# ex 5 - 2
# 코드에 대한 설명 파일을 불러보자
code_data = pd.read_csv('./datas/code_pre.csv', encoding = 'cp949')
# 코드에 대한 설명을 확인하자
print("code_data['속성명'] >>", code_data['속성명'])

# 분석 대상을 정의하자
features = [code_data['속성명'][i] for i in [4,5,6,7,9]]
print("features >>", features)

# 필요한 데이터 컬럼의 추출
rdata = data.iloc[:,[4,5,6,7,9]]
x = np.array(rdata)
n, p = x.shape

# 데이터정규화
from sklearn.preprocessing import StandardScaler 
x = StandardScaler().fit_transform(x) 
x.mean(axis=0)
x.std(axis=0)
x
```

```python

# ex 5 - 3

import matplotlib.pyplot as plt

bvec = np.random.uniform(size = p)
bvec = bvec / np.linalg.norm(bvec) # 유클리드 거리? 
# (x[0,:]*bvec).sum()
y = x@bvec # y는 지수, 
# 좋은 지수는 bvec의 숫자를 고르는 것?, 좋은 지수는 사람별로 확 차이가 나는 것, 즉 차이(여기서는 "분산")가 어떻게 나는지 보여주는 지수가 좋은 지수다 
# 그래프로 가장 많이 벌어지게 되는 것 -> 그게 바로 1 주성분이다 -> 수학적으로 계산이 된다!! (스코어로)

plt.hist(y, bins = 200, range = (-10,10))
```

```python

# ex 5 - 4, 
# 지수, 주성분 백테로 표현이 된 데이터

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(x)
print("pca.explained_variance_ratio_ >>", pca.explained_variance_ratio_)
print("pca.singular_values_ >>", pca.singular_values_)
print("pca.components_ >>", pca.components_)
print(features) # pca.components_ 의 array와 features의 동일 index 위치에서 값의 영향 범위체크, 어떤 값이 지수에 어떤 영향을 주는지

pc_score = pca.fit_transform(x)
print("pca.fit_transform(x) >>", pc_score) # 한 사람당 지수가 2개
print("x.shape >>", x.shape)
# pc_score 로 그래프가 "나누어 진다" 면, 그렇게 집합을 묶을 수 있다, 완전 잘 군집된 것
# 이런 방법이 차원 축소 

plt.plot(pc_score[:,0], pc_score[:,1], '.') 
```

```python

# ex 6, SVD
# Singular Value Decomposition - UDVT

# care about full_matrices = False
u, s, vh = np.linalg.svd(x, full_matrices=False) # svd 분해하기임
print("u@np.diag(s)@vh >>", u@np.diag(s)@vh) # 얘기 x랑 같아야 함
print("위랑 같이야 할 x >>", x)
print("u.shape >>", u.shape)
print("s.shape >>", s.shape)
print("vh.shape >>", vh.shape) # vh의 행을 보면 지수 만들기를 알 수 있다
print("vh의 행을 보면 지수 만들기를 알 수 있다, vh >>", vh)
# vh[1,:] # 2번 지수...
print("중요도?! - explained_variance_ratio_ 인 s >> ", s**2/np.sum(s**2)) # 중요도?! - explained_variance_ratio_

print("vh[0,:]*vh[1,:]).sum() >>", (vh[0,:]*vh[1,:]).sum())



# UDV^T 에서 V^T matrix를 확인
# V행렬의 열벡터는 직교함
# plt.plot(u[:,0],u[:,1], '.')


# considering the size s
plt.plot(s[0]*u[:,0],s[1]*u[:,1], '.')
```
